# a)

```python
import numpy as np
import matplotlib.pyplot as plt

def P(t) :
  return 18 + ((24*t)/(t+2))

t = np.linspace (0.01, 30, 100)

potencia = P(t)

plt.figure
plt.plot (t, potencia)
plt.xlabel ("Tempo min")
plt.ylabel ("Potência kW")
plt.title ("Potência ao longo do tempo")

plt.show()
```

# b)

```python
import sympy as sp

t = sp.symbols('t', positive=True)
P = 18 + (24*t)/(t + 2)
infinito = "∞"

limite_zero = sp.limit(P, t, 0, dir="+")
limite_infinito = sp.limit(P, t, sp.oo)

print("Limite quando t -> 0+:", limite_zero)
print("Limite quando t -> ", infinito, ":", limite_infinito)
```

# c)

```python
def P(t) :
  return 18 + ((24*t)/(t+2))

variacao_media = (P(3) - P(1)) / (3-1)

print("P(1) =", P(1), "kW")
print("P(3) =", P(3), "kW")
print("Variação média =", variacao_media, "kW/min")
```

# d)

```python
import sympy as sp

t = sp.symbols('t', positive=True)
P = 18 + (24*t)/(t + 2)

derivada = sp.diff(P, t)

derivada_simplificada = sp.simplify(derivada)

print("Derivada simplificada: P'(t) =", derivada_simplificada)
```

# e)

```python
# Derivada
def P_linha(t):
    return 48 / (t + 2)**2

# Avaliações
print("P'(1) =", P_linha(1))
print("P'(6) =", P_linha(6))
```

# f)

A derivada P' (t) é positiva para t>0, indicando que a potência está sempre aumentando. Com isso, é possível concluir que se os valores da derivada diminuem ao longo do tempo, o crescimento da potência ocorre de forma cada vez mais lenta, aproximando-se de um valor limite.
