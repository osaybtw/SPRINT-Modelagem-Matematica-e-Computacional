# 1a.

```python
import numpy as np
import matplotlib.pyplot as plt

def L(p) :
  return -35*p**2 + 298*5*p -406

p = np.linspace (1.5, 6.5, 100)

plt.plot(p, L(p))
plt.title("Gráfico da função L(p)")
plt.xlabel("Tarifa: p")
plt.ylabel("Lucro Diário: L(p)")
plt.grid()

plt.show()
```

# 2b.

```python
import sympy as sp

p = sp.Symbol('p')
L = -35*p**2 + 298*5*p -406
derivada = sp.diff(L, p)

print("Função lucro: ", L)
print("Derivada da função lucro:", derivada)
```

# 3c.

```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt

L = -35*p**2 + 298.5*p - 406
p = sp.Symbol('p')
derivada = sp.diff(L, p)
ponto_crit = sp.solve(derivada, p)

print("Derivada da função: ", derivada)
print(f"Ponto crítico: {float(ponto_crit[0]):.2f}")

valores_p = np.linspace(1.5, 6.5, 100)
valores_derivada = -70*valores_p + 298.5

plt.plot(valores_p, valores_derivada)
plt.scatter(float(ponto_crit[0]), 0)
plt.title("Gráfico da derivada L'(p)")
plt.xlabel("Tarifa (p)")
plt.ylabel("L'(p)")
plt.grid()

plt.show()
```

# 4d.

```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt

L = -35*p**2 + 298.5*p - 406
p = sp.Symbol('p')
L2 = sp.diff(L, p, 2)

print("Segunda derivada: ", L2)

x = np.linspace(1.5, 6.5, 100)
y = np.full_like(x, -70)

plt.plot(x, y)
plt.title("Gráfico de da Segunda Derivada")
plt.xlabel("p")
plt.ylabel("L''(p)")
plt.grid()

plt.show()
```

# 5e.

```python
import sympy as sp

L = -35*p**2 + 298.5*p - 406
p = sp.Symbol('p')
L1 = sp.diff(L, p)
L2 = sp.diff(L, p, 2)
crit = sp.solve(L1, p)[0]
valor_crit = L.subs(p, crit)
valor_inicio = L.subs(p, 1.5)
valor_fim = L.subs(p, 6.5)

print(f"Ponto crítico: {float(crit):.2f}")
print(f"Segunda derivada: {float(L2):.2f}")
print(f"Lucro no ponto crítico: {float(valor_crit):.2f}")
print(f"Lucro em p = 1.5: {float(valor_inicio):.2f}")
print(f"Lucro em p = 6.5: {float(valor_fim):.2f}")
```
